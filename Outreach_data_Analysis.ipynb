```python
import pandas as pd

outreach_df = pd.read_excel('Outreach Data.xlsx')
```

```python
#number of rows and columns in the dataset
print('Number of rows:', outreach_df.shape[0])
print('Number of columns:', outreach_df.shape[1])
```

```python
#Checking the data types of the variables in the data
print(outreach_df.dtypes)
```

```python
#Getting the summary statistics of the numeric variables -  central tendency, dispersion, and shape of the distribution of each variable.
print(outreach_df.describe())
```

```python
#Getting the sum of the individual columns

column_sums = outreach_df.sum()

# Print the column sums
print(column_sums)
```

```python
#The Health facilities with the highest number of children with SAM

import numpy as np
import matplotlib.pyplot as plt
#create a subset of the data
subset = outreach_df.loc[:, ['Facility Name', 'Number of Children with SAM']]

#Grouping the data
grouped = subset.groupby('Facility Name').mean()

#Sorting the data in descending order based on the mean number of children screened for malnutrition
sorted_data = grouped.sort_values(by='Number of Children with SAM', ascending=False)

#Creating a bar chart of the top 15 health facilities with the highest mean number of children screened for malnutrition

top10 = sorted_data.head(15)
top10.plot(kind='bar', y= 'Number of Children with SAM', legend=False)
plt.title('Facilities Name with the Highest Mean Number of Children with SAM')
plt.ylabel('Mean Number of Children Screened')
plt.show()

```

```python
#The Health facilities with the highest number of children at Risk for Malnutrition

#create a subset of the data
subset = outreach_df.loc[:, ['Facility Name', 'Number of children under MAM']]

#Grouping the data
grouped = subset.groupby('Facility Name').mean()

#Sorting the data in descending order based on the mean number of children at Risk for Malnutrition
sorted_data = grouped.sort_values(by='Number of children under MAM', ascending=False)

#Creating a bar chart of the top 15 health facilities with the highest mean number of children at Risk for Malnutrition

top10 = sorted_data.head(15)
top10.plot(kind='bar', y= 'Number of children under MAM', legend=False)
plt.title('Facilities Name with the Highest Mean Number of children under MAM')
plt.ylabel('Mean Number of children under MAM')
plt.show()
```

```python
#The Health facilities with the highest number of children Malnourished

#create a subset of the data
subset = outreach_df.loc[:, ['Facility Name', 'Number of children Malnourished']]

#Grouping the data
grouped = subset.groupby('Facility Name').mean()

#Sorting the data in descending order based on the mean number of children with MUAC above 13.5
sorted_data = grouped.sort_values(by='Number of children Malnourished', ascending=False)

#Creating a bar chart of the top 15 health facilities with the highest mean number of children with MUAC above 13.5

top10 = sorted_data.head(15)
top10.plot(kind='bar', y= 'Number of children Malnourished', legend=False)
plt.title('Facilities Name with the Highest Number of children Malnourished')
plt.ylabel('Mean Number of children Malnourished')
plt.show()
```

```python
#The Health facilities with the highest Total malnutrition(Children and Pregnant and Lactating women)

#create a subset of the data
subset = outreach_df.loc[:, ['Facility Name', 'Total Number of Malnutrition']]

#Grouping the data
grouped = subset.groupby('Facility Name').mean()

#Sorting the data in descending order based on the mean of the Total Number of Malnutrition
sorted_data = grouped.sort_values(by='Total Number of Malnutrition', ascending=False)

#Creating a bar chart of the top 15 health facilities with the highest Total Number of Malnutrition

top10 = sorted_data.head(15)
top10.plot(kind='bar', y= 'Total Number of Malnutrition', legend=False)
plt.title('Facilities Name with the Highest Total Number of Malnutrition')
plt.ylabel('Mean Total Number of Malnutrition')
plt.show()
```

```python
#The Health facilities with the highest number of children with MUAC above 13.5

#create a subset of the data
subset = outreach_df.loc[:, ['Facility Name', 'Number of children with MUAC above 13.5']]

#Grouping the data
grouped = subset.groupby('Facility Name').mean()

#Sorting the data in descending order based on the mean number of children with MUAC above 13.5
sorted_data = grouped.sort_values(by='Number of children with MUAC above 13.5', ascending=False)

#Creating a bar chart of the top 10 health facilities with the highest mean number of children with MUAC above 13.5

top10 = sorted_data.head(10)
top10.plot(kind='bar', y= 'Number of children with MUAC above 13.5', legend=False)
plt.title('Facilities Name with the Highest Mean Number of children with MUAC above 13.5')
plt.ylabel('Mean Number of Children screened')
plt.show()
```

```python
#The Health facilities with the highest Number of PLW with MUAC <21cm - This represents the number of pregnant and Lactating women with malnutrition

#create a subset of the data
subset = outreach_df.loc[:, ['Facility Name', 'Number of PLW with MUAC <21cm']]

#Grouping the data
grouped = subset.groupby('Facility Name').mean()

#Sorting the data in descending order based on the mean number of PLW with malnutrition
sorted_data = grouped.sort_values(by='Number of PLW with MUAC <21cm', ascending=False)

#Creating a bar chart of the top 15 health facilities with the highest mean number of PLW with malnutrition

top10 = sorted_data.head(15)
top10.plot(kind='bar', y= 'Number of PLW with MUAC <21cm', legend=False)
plt.title('Facilities Name with the Highest Mean Number of PLW with MUAC <21cm')
plt.ylabel('Mean Number of PLW with MUAC <21cm')
plt.show()
```

```python
#Identifying the correlation using scatterplots to visualize the relationship between the number of PLW with MUAC <21cm and the number of children malnourished.
outreach_df.plot.scatter(x='Number of PLW with MUAC <21cm', y='Number of children Malnourished')
plt.show()
```

```python
#plot a boxplot of malnutrition cases for wards

# set figure size and font size
plt.figure(figsize=(8,6))
plt.rcParams.update({'font.size': 8})

# create box plot
outreach_df.boxplot(column='Total Number of Malnutrition', by='Ward', vert=False, figsize=(8,6), widths=0.6, patch_artist=True)

# set label properties
plt.xticks(rotation=0, fontsize=12)
plt.title('The Distribution of the Total Number of Malnutrition by Ward')
plt.xlabel('Total Number of Malnutrition')
plt.ylabel('Ward')

plt.show()
```

```python
#Creating a Frequency Table of referrals for Number of children referred for OTP/SFP

# Group the data by health facility and count the number of referrals
referral_counts = outreach_df.groupby('Facility Name')['Number of children reffered for OTP/SFP'].count()

# Create a frequency table
freq_table = pd.DataFrame({'Facility Name': referral_counts.index, 'Referral Count': referral_counts.values})

# Print the frequency table
print(freq_table)

#plotting a frequency table for the number of children reffered for OTP/SFP
freq_table.plot.bar()
plt.show()
```
